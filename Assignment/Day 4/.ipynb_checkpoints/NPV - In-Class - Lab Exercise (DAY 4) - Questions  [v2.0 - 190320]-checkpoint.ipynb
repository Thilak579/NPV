```python
w=0.2
Year=[2001,2002,2003,2004,2005]
Export=[42.5,44.5,48.3,57.24,69.18]
Import=[54.4,53.8,61.6,74.15,89.33]
b1=np.arange(len(Year))
b2=[i+w for i in b1]
plt.bar(b1,Import,w,label='Import')
plt.bar(b2,Export,w,label='Export')
plt.xticks(b1+w/2,Year)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Data')
plt.legend()
plt.show()
```
